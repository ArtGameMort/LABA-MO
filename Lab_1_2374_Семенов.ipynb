Для первой лабораторной работы я выбрал датасет, содержащий реальные данные о ценах акций компании Apple. Этот набор данных охватывает фондовый рынок и цены акций различных компаний за определённый период, и вся информация была собрана с сайта: 
https://www.kaggle.com/datasets/aayushmishra1512/faang-complete-stock-data/data?select=Apple.csv

Задача: Проанализировать финансовые данные компании Apple, в частности изменения цен акций, за определённый период времени.

Описание датасета:
Атрибуты:
- Open - цена акций в начале торгового дня (числовой тип данных)

Среднее значение и Стандартное отклонение:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Apple.csv')
column = 'Open'
    
mean_value = data[column].mean()
std_value = data[column].std()
print(f"Среднее значение для {column}: {mean_value}")
print(f"Стандартное отклонение для {column}: {std_value}")
```

Гистограмма:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Apple.csv')
column = 'Open'
    
mean_value = data[column].mean()
std_value = data[column].std()
    
plt.figure(figsize=(10, 6))
sns.histplot(data[column], kde=True)
plt.title(f'Гистограмма для {column}')
plt.show()
    
missing_values = data[column].isnull().sum()
print(f"Пропущенные значения для {column}: {missing_values}")
    
if missing_values > 0:
    print(f"Заполнение пропущенных значений в {column} средним")
    data[column] = data[column].fillna(mean_value)
    
Q1 = data[column].quantile(0.25)
Q3 = data[column].quantile(0.75)
IQR = Q3 - Q1
outliers = ((data[column] < (Q1 - 1.5 * IQR)) | (data[column] > (Q3 + 1.5 * IQR))).sum()
print(f"Количество выбросов в {column}: {outliers}")
```

В данном датасете есть пропущенных значений.

Для заполнения пропусков стоит:
 - Стоит заполнять пропуски средним значением. 
 - Заполнить среднеквадратичным (медианным) значением.
 - Восстановление данных по закону распределения.
 - Удалить это значение, но такой метод подходит, если у нас большое кол-во данныx, а также малое кол-во пропусков.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('Apple.csv')
data = data.drop(columns=['Date'])  

correlation_matrix = data.corr()

print("Корреляционная матрица:")
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Корреляционная матрица между атрибутами')
plt.show()

sns.pairplot(data)
plt.show()

high_corr = correlation_matrix[(correlation_matrix > 0.8) | (correlation_matrix < -0.8)]
print("Высококоррелированные атрибуты:")
print(high_corr)
```

В данном шаге мы построили корреляционную матрицу между атрибутами и нашли высококоррелированные атрибуты нашего датасета

Из последней матрицы можно заметить, что все атрибуты, за исключением Volume (количество проданных акций за день), имеют высокую корреляцию. Это вполне ожидаемо, поскольку мы анализируем различные пиковые цены акций одной компании в течение одного дня, и каждый атрибут будет сильно зависеть от других (например, цена закрытия не может не зависеть от цены открытия). Таким образом, все значения демонстрируют высокую корреляцию, в то время как значение Volume вообще не показывает корреляции.

Вывод: Мы определили средние показатели для каждого атрибута. Построили гистограммы для каждого атрибута. Выяснили, что многие параметры, представленные в датасете, имеют высокую корреляцию. Нашли пропущенные значения и представили способы их заполнения.
